```python
# all these below import packages are been installed in pyproject.toml file when we ran the uv add ...
from langchain_groq import ChatGroq
llm = ChatGroq(model="llama-3.1-8b-instant", temperature=0)
```

```python
from langchain_community.document_loaders import PyPDFLoader

pdf_path = "./Docs/fabric-admin.pdf"

loader = PyPDFLoader(pdf_path)
docs = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = text_splitter.split_documents(docs)
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
embedding_model = GoogleGenerativeAIEmbeddings(
    model="models/gemini-embedding-001"
)
```

```python
# add the new pdf into existing vectorstore Vector ( means rag.pdf is added into vectorstore , so we will add the other 2 pdf into the same vectorstore )
from langchain_community.vectorstores import FAISS # FAISS DB is a free vector store
vectorstore = FAISS.load_local("./Vector/", embedding_model, allow_dangerous_deserialization=True)
```

```python
vectorstore.add_documents(chunks)
vectorstore.save_local("./Vector/")
```
